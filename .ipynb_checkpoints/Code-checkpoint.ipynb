```python
array1 = ttv.astype(float)
array2 = container.astype(float)
array3 = emc.astype(float)

def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    similarity = dot_product / (norm_vec1 * norm_vec2)
    return similarity

# 計算三組陣列的餘弦相似度
similarity_12 = cosine_similarity(array1, array2)
similarity_13 = cosine_similarity(array1, array3)
similarity_23 = cosine_similarity(array2, array3)

print("餘弦相似度 (array1 和 array2):", similarity_12)
print("餘弦相似度 (array1 和 array3):", similarity_13)
print("餘弦相似度 (array2 和 array3):", similarity_23)
```
